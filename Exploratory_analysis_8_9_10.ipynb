```python
import numpy as np
from scipy.spatial.distance import euclidean
import pandas as pd

def dtw_distance(series1, series2):

    length1 = len(series1)
    length2 = len(series2)

    cost = np.full((length1 + 1, length2 + 1), np.inf)


    cost[0, 0] = 0


    for i in range(1, length1 + 1):
        for j in range(1, length2 + 1):

            dist = euclidean(series1[i - 1], series2[j - 1])


            cost[i, j] = dist + min(cost[i - 1, j], cost[i, j - 1], cost[i - 1, j - 1])


    return cost[length1, length2]


series1 = data_9_2Mar
series2 = data_10_2Mar
similarity_score = dtw_distance(series1, series2)
print(f"DTW similarity score: {similarity_score}")
```
